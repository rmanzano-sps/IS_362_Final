```python
import pandas as pd
import numpy as np

df = pd.read_csv('scores.csv')
df
```

```python
scores_df = pd.DataFrame()
scores_df['School Name'] = df['School Name']
scores_df['Borough'] = df['Borough']
scores_df['Zip Code'] = df['Zip Code']
scores_df['Average Score (SAT Math)'] = df['Average Score (SAT Math)']
scores_df['Average Score (SAT Writing)'] = df['Average Score (SAT Writing)']
scores_df['Average Score (SAT Writing)'] = df['Average Score (SAT Writing)']
scores_df['Percent Tested'] = df['Percent Tested']


scores_df = scores_df[scores_df['Percent Tested'].notnull()]
scores_df = scores_df[scores_df['Average Score (SAT Math)'].notnull()]
scores_df = scores_df[scores_df['Average Score (SAT Writing)'].notnull()]
scores_df = scores_df.reset_index(drop=True)
scores_df
```
