```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim

df = pd.read_csv("NYPD_Complaint_Data_Historic.csv", low_memory=False)

df = df.head(500)
df
```

```python
crimes_df = pd.DataFrame()
crimes_df['Latitude'] = df['Latitude']
crimes_df['Longitude'] = df['Longitude']
crimes_df['BORO_NM'] = df['BORO_NM']
crimes_df['LAW_CAT_CD'] = df['LAW_CAT_CD']
crimes_df['PD_DESC'] = df['PD_DESC']
crimes_df['OFNS_DESC'] = df['OFNS_DESC']
crimes_df = crimes_df[crimes_df['Latitude'].notnull()]
crimes_df = crimes_df[crimes_df['Longitude'].notnull()]
crimes_df = crimes_df[crimes_df['PD_DESC'].notnull()]
crimes_df = crimes_df.reset_index(drop=True)
crimes_df
```

```python
geolocator = Nominatim(user_agent="crimes_df")

for index, row in crimes_df.iterrows():
    coordinates = str(row['Latitude']) + ', ' + str(row['Longitude'])
    location = geolocator.reverse(coordinates)
    crimes_df.at[index, 'postcode'] = location.raw['address']['postcode']

crimes_df_final = crimes_df
crimes_df_final
```

```python
crimes_df_copy = crimes_df_final
crimes_df_2 = pd.DataFrame()
crimes_df_2['Post_Code'] = crimes_df_copy['postcode']
crimes_df_2['LAW_CAT_CD'] = crimes_df_copy['LAW_CAT_CD']

# crimes_df_2['FELONY'] = crimes_df_copy.groupby('postcode').groupby('LAW_CAT_CD').size()
# crimes_df_2
crimes_df_2 = crimes_df_2.Post_Code.value_counts().reset_index(name='Crimes')
crimes_df_2 = crimes_df_2.rename(columns={'index':'Zip Codes'})

crimes_df_2
```
